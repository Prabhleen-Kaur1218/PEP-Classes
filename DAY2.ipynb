```python
#Write a Python Program to Check Armstrong Number?

for num in (range(10000)):
    strnum = str(num)
    pow = len(strnum)
    sum = 0
    for i in (range(len(strnum)-1)):
        sum += int(strnum[i])**pow
    if(sum == int(strnum)):
        print(int(strnum))
    
```

```python
#Write a python program to find the HCF?
num1 = int(input("Num1"))
num2 = int(input("Num2"))

smaller = 0
if num1<num2:
    smaller = num1
else:
    smaller = num2
for i in range(1, smaller+1):
    if(smaller%i == 0):
        hcf = i
        
print(f'HCF for {num1} and {num2} is {hcf}')
```
