```python
list=[32,31,40,54,32,54,65,63,67,87]

#Write a code to sort the above list

for i in range(len(list)-1):
    for j in range(i+1,len(list)):
        if list[i]>list[j]:
            list[i],list[j] = list[j], list[i]

print(list)
```
