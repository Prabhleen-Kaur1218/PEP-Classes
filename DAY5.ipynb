```python
def plus(x,y):
    sum = x+y
    return sum

def minus(c,d):
    diff = c-d
    return diff

def func3(a,b):
    frm1 = plus(a,b)/minus(a,b)
    return frm1

func3(a=65,b=66)
```

```python
def fact(n):
	if n == 0:
		return 1
	return n * fact(n-1)
	
print(fact(10))
```

```python
def isPrime(x):
    for i in (range(2,x)):
        if x%i==0:
            return f"{x} is not prime"
            break
        
        else:
            return f"{x} is a prime number"

isPrime(x=7)
```

```python
def fibo(n):
    if n<= 1:
        return n
    else :
        return(fibo(n-1)+fibo(n-2))

nterms = int(input("Number"))

print("fibonacci series")
for i in range(nterms):
    print(fibo(i))
```

```python
def armstrongnum(a):
    stra = str(a)
    power = len(stra)
    sum = 0
    for i in range(len(stra)-1):
        sum += int(stra[i])**power
    
    if sum == a:
        return f'{a} is an Armstrong number'
    else:
        return f'{a} is not an Armstrong number'
        
```

```python
armstrongnum(370)
```
